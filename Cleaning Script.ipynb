```python
import pandas as pd

# Load the Excel file
file_path = 'C:/Users/abdulrahman/OneDrive/Desktop/Final project/Data xlsx.xlsx'

# Inspect the sheet names first
excel_file = pd.ExcelFile(file_path)
excel_file.sheet_names
```

```python
# Load the second sheet 'DATOS' into a DataFrame for inspection
df_datos = pd.read_excel(file_path, sheet_name='DATOS')

# Display the first few rows of the dataset to understand its structure
df_datos.head()
```

```python
# Check for missing values in each column
missing_values = df_datos.isnull().sum()

# Display missing values
missing_values
```

```python
# Drop columns that have excessive missing values and are not necessary for analysis
columns_to_drop = ['SIGLAS', 'LINK', 'ARCHIVO', 'UBICACIÓN DE INFORMACIÓN EN EL DOCUMENTO', 'OBSERVACIONES', 
                   'NIVEL DE DESAGREGACIÓN', 'INFORMACÓN', 'FUENTE','CALCULO','CÓDIGO','PAIS','REGIÓN','INDICADOR',
                  'UNIDAD DEL INDICADOR']

df_cleaned = df_datos.drop(columns=columns_to_drop)

# Drop rows with missing values in 'VALOR' as they are critical for analysis
df_cleaned = df_cleaned.dropna(subset=['VALOR'])

# Ensure 'AÑO' is treated as integer and 'VALOR' as float
df_cleaned['AÑO'] = df_cleaned['AÑO'].astype(int)
df_cleaned['VALOR'] = pd.to_numeric(df_cleaned['VALOR'], errors='coerce')

# Rename columns for easier use
df_cleaned = df_cleaned.rename(columns={
    'AÑO': 'Year',
    'VALOR': 'Value'
})

# Display the first few rows
df_cleaned.head()
```

```python
df_cleaned.to_csv('C:/Users/abdulrahman/OneDrive/Desktop/Final project/cleaned_data.csv', index=False)
```
